```python
# -------------------------------------------------------------
# 작성자 : 백강민
# 작성목적 : SKALA Python Day1 - generator 실습 코드 1
# 작성일 : 2025-01-12
# 변경사항 내역 : 
#   2025-01-12 - 최초 작성
# -------------------------------------------------------------

import sys

print("【 실습 과제 1 – 일반 리스트 vs. 제너레이터 】\n")

# =========================
# 1. 리스트 방식
# =========================
print("1. 0부터 999,999까지의 정수를 담는 리스트를 생성하고 총합 구하기")

numbers_list = list(range(1_000_000))
list_sum = sum(numbers_list)

print(f"   - 총합: {list_sum}")


# =========================
# 2. 제너레이터 방식
# =========================
print("2. 같은 결과를 제너레이터 함수로 구현")

def number_generator(limit):
    for i in range(limit):
        yield i

numbers_gen = number_generator(1_000_000)
gen_sum = sum(numbers_gen)

print(f"   - 총합: {gen_sum}")


# =========================
# 3. 메모리 사용 차이 비교
# =========================
print("3. 두 방법의 메모리 사용 차이를 sys.getsizeof()로 확인")

memory_diff = sys.getsizeof(numbers_list) - sys.getsizeof(numbers_gen)

print(f"   - 리스트 방식 메모리: {sys.getsizeof(numbers_list):,} bytes")
print(f"   - 제너레이터 방식 메모리: {sys.getsizeof(numbers_gen):,} bytes")
print(f"   - 메모리 차이: {memory_diff:,} bytes")
```
