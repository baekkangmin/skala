```python
class BankAccount:
    def __init__(self, owner, balance=0):
        self.owner = owner  # 계좌 주인
        self.balance = balance  # 계좌 잔액
    
    def deposit(self, amount):  # 입금 메서드
        self.balance += amount
        print(f"{amount}원이 입금되었습니다. 잔액: {self.balance}원")
    
    def withdraw(self, amount):  # 출금 메서드
        if amount > self.balance:
            print("잔액이 부족합니다!")
        else:
            self.balance -= amount
            print(f"{amount}원이 출금되었습니다. 잔액: {self.balance}원")
```
